The variable hashtag stores the value of hashtag, which will be used to search for tweets. 

```r
hashtag="#superbowl";
```

The function serachTwitter() is used to serach for tweets based on passed arguments.
Here we are passing only two arguments - search string and n.
The variable hashtag which stores the string "#superbowl" is passed as search string.
n specifies the maximum number of tweets to return.
Here we are passing n=2500.
The returned tweets in form of a list are stored in the variable tweets.

```r
tweets <- searchTwitter(hashtag,n=2500)
```

The function twListToDF() is used to convert the returned list of tweets into a data frame.
The returned data frame is stored in a variable df.

```r
df=twListToDF(tweets)
Sys.setlocale(locale="C")
```

The function lookupUsers() is used to get user information.
df$screenName will return the column 'screenName' of the data frame df as a list, this list will contain screenNames of the users corresponding to the collected tweets.
This list of screen names is then passed to the function lookupUsers() to get the information corresponding to those users.
Again the function lookupUsers will return the users info in form of a list, which is converted into a data frame using the function twListToDF(). 
Now df1 contains users information.

```r
df1=twListToDF(lookupUsers(df$screenName))
```

Installing and using package 'dplyr'to manipulate the data frame df1

```r
install.packages("dplyr")
library("dplyr")
```

The function filter() is used to eliminate rows from df1, where location field is empty.

```r
df2=filter(df1,location!="")
```

Installing and using the following packages to draw and plot a map - 'ggmap', 'maps', 'mapproj', 'ggplot2', and 'maptools'.

The function write() is used to write the data frame df2 into a csv file.

```r
write.csv(df2,file="location.csv")
```

The function read() is used to read the data from the csv file.
Returned data is stored in the variable temp in form of a data frame.

```r
temp=read.csv(file="location.csv",head=TRUE,sep=",")
```

The location column of the data frame 'temp', is passed to the function table(), which will append the corresponding frequencies of the locations.
The returned data is converted into a data frame and stored in test1.

```r
test1=as.data.frame(table(temp$location))
```

The 'Var1' column of test1, which contains the locations, is passed to the function geocode(), which will return a dataframe containing the longitude and latitute corresponding to the locations.  

```r
test=geocode(as.character(test1$Var1))
test1
test
```

The function cbind() is used to concatenate data frames test1 and test.

```r
compliment=cbind(test1,test)
compliment
```

The function get_map() is used to get a map of united states.

```r
mymap=get_map(source="google", zoom=4, scale=2, location="united states")
```

The following line of code will take data from the data frame 'compliment' in order to plot the map.

```r
ggmap(mymap) + geom_point(data = compliment, aes(x = lon, y = lat, colour="red", size=Freq))
```
